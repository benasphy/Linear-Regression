```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

study = np.array([3,5,6,8,10])
score = np.array([60,75,80,86,91])

X = study.reshape(-1,1)
y = score

model = LinearRegression()

model.fit(X,y)

new_study = np.array([7]).reshape(-1,1)
predictions = model.predict(new_study)

plt.plot(study, model.predict(X), color = 'red', label = 'Regression Line')

for studyhour, prediction in zip(new_study.flatten(), predictions):
  print(f"Predicted Score for a {studyhour} hours is {prediction:.2f}")
```
